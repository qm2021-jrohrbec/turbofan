Weibull distribution density:

$ f_{\lambda,k}(x) =
\begin{cases}
\frac{k}{\lambda}\left(\frac{x}{\lambda}\right)^{k-1}e^{-(x/\lambda)^{k}} &, x\geq0 \\
0 &, x<0
\end{cases}$.

$k$ is also called the shape parameter and $\lambda$ the scale parameter.

```python
def weibull_density(x, shape, location, scale):
    '''
    Only x >= 0
    '''
    return (shape/scale) * (((x-location)/scale) ** (shape - 1)) * np.exp(-(((x-location)/scale) ** shape))
```
