```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0., 5., 0.01)
```

```python
t2 = t
t2[0] = 0.001
```

# Distributions
## 1. Exponential Distribution

```python
def exp_rf(t, l):
    return np.exp(- l * t)
```

```python
l1 = 0.5
l2 = 1.
l3 = 2.
```

```python
rf1 = exp_rf(t, l1)
rf2 = exp_rf(t, l2)
rf3 = exp_rf(t, l3)
```

#### Density Plot

```python
fig, ax = plt.subplots()
ax.plot(t, l1*rf1, label=r'$\lambda_1 = 0.5$')
ax.plot(t, l2*rf2, '--', label=r'$\lambda_2 = 1$')
ax.plot(t, l3*rf3, '-.', label=r'$\lambda_3 = 2$')
ax.set_ylabel(r'Density ($\it{pdf}$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
#plt.show()
```

#### Reliability Function Plot

```python
fig, ax = plt.subplots()
ax.plot(t, rf1, label=r'$\lambda_1 = 0.5$')
ax.plot(t, rf2, '--', label=r'$\lambda_2 = 1$')
ax.plot(t, rf3, '-.', label=r'$\lambda_3 = 2$')
ax.set_ylabel(r'Reliability Funciton ($\it{rf}$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
#plt.show()
```

#### Hazard Rate Plot

```python
fig, ax = plt.subplots()
ax.plot(t, np.repeat(l1, len(t)), label=r'$\lambda_1 = 0.5$')
ax.plot(t, np.repeat(l2, len(t)), '--', label=r'$\lambda_2 = 1$')
ax.plot(t, np.repeat(l3, len(t)), '-.', label=r'$\lambda_3 = 2$')
ax.set_ylabel(r'Hazard Function ($\it{hf}$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
#plt.show()
```

## 2. Weibull

```python
def weib_rf(t, mu, beta):
    return np.exp(-((mu*t)**beta))

def weib_hf(t,mu,beta):
    return beta*mu*((mu*t)**(beta-1))
```

```python
eta1 = 1
eta2 = 2
beta1 = 0.5
beta2 = 1
beta3 = 1.5
beta4 = 3
```

```python
rf11 = weib_rf(t, eta1, beta1)
rf12 = weib_rf(t, eta1, beta2)
rf13 = weib_rf(t, eta1, beta3)
rf14 = weib_rf(t, eta1, beta4)

rf21 = weib_rf(t, eta2, beta1)
rf22 = weib_rf(t, eta2, beta2)
rf23 = weib_rf(t, eta2, beta3)
rf24 = weib_rf(t, eta2, beta4)
```

```python
fig, ax = plt.subplots()
ax.plot(t, rf11, label=r'$\beta_1 = 0.5$')
ax.plot(t, rf12, '--', label=r'$\beta_2 = 1$')
ax.plot(t, rf13, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, rf14, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Reliability Function ($\it{rf}$, $\eta = 1$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
#plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, rf21, label=r'$\beta_1 = 0.5$')
ax.plot(t, rf22, '--', label=r'$\beta_2 = 1$')
ax.plot(t, rf23, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, rf24, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Reliability Function ($\it{rf}$, $\eta = 2$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
#plt.show()
```

```python
hf11 = weib_hf(t2, eta1, beta1)
hf12 = weib_hf(t, eta1, beta2)
hf13 = weib_hf(t, eta1, beta3)
hf14 = weib_hf(t, eta1, beta4)

hf21 = weib_hf(t2, eta2, beta1)
hf22 = weib_hf(t, eta2, beta2)
hf23 = weib_hf(t, eta2, beta3)
hf24 = weib_hf(t, eta2, beta4)
```

```python
fig, ax = plt.subplots()
ax.plot(t, hf11, label=r'$\beta_1 = 0.5$')
ax.plot(t, hf12, '--', label=r'$\beta_2 = 1$')
ax.plot(t, hf13, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, hf14, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Hazard Function ($\it{hf}$, $\eta = 1$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
ax.set_ylim(0, 10)
#plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, hf21, label=r'$\beta_1 = 0.5$')
ax.plot(t, hf22, '--', label=r'$\beta_2 = 1$')
ax.plot(t, hf23, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, hf24, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Hazard Function ($\it{hf}$, $\eta = 2$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
ax.set_ylim(0, 10)
#plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, hf11 * rf11, label=r'$\beta_1 = 0.5$')
ax.plot(t, hf12 * rf12, '--', label=r'$\beta_2 = 1$')
ax.plot(t, hf13 * rf12, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, hf14 * rf14, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Density Function ($\it{df}$, $\eta = 1$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
ax.set_ylim(0, 1.2)
#plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t, hf21 * rf21, label=r'$\beta_1 = 0.5$')
ax.plot(t, hf22 * rf22, '--', label=r'$\beta_2 = 1$')
ax.plot(t, hf23 * rf22, '-.', label=r'$\beta_3 = 1.5$')
ax.plot(t, hf24 * rf24, '-.', label=r'$\beta_4 = 3$')
ax.set_ylabel(r'Density Function ($\it{df}$, $\eta = 2$)')
ax.set_xlabel(r't')
ax.legend()
ax.set_xlim(0, t.max())
ax.set_ylim(0, 2.5)
#plt.show()
```
